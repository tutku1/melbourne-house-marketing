```python
#converting objects to categorical variable
object_columns = df.select_dtypes("object").columns

#car and bathroom are integer ut also categorical values
object_columns = np.append(object_columns,"Car")
object_columns = np.append(object_columns,"Bathroom")

print(object_columns)
```

```python
for colname in object_columns:
    df[colname] = df[colname].astype("category")
    
    df.info()
```

```python
#removing duplicates
df.drop_duplicates(inplace=True, keep="first")
df.duplicated().sum()
```

```python
#checking min, max and mean of values to ensure it makes sense
df.describe().T
```

```python
#landsize and Building Area can't be 0 or negative. Converting them to nan
df.loc[df["Landsize"] <= 0, "Landsize"] = np.nan
df.loc[df["BuildingArea"] <= 0, "BuildingArea"] = np.nan
```

```python
#removing false Year Built because Melbourne Founded 1835
df.drop(df.loc[df["YearBuilt"] < 1835].index, inplace=True)
```

```python
#adding features for house age
df["Age"] = 2022 -df["YearBuilt"]
df.drop(df.loc[df["Age"] < 0].index, inplace=True)
```

```python
df.head()
```

```python
df = df.drop(["YearBuilt"], axis = 1)
```

```python
df.head()
```

```python
numerical_columns = df.select_dtypes("float64", "int64").columns # finds the integer and float columns in the datasets
numerical_columns
```

```python
num_cols = len(numerical_columns)
num_rows = (num_cols + 1) // 2  # Create a 2-column grid

# Creating the subplots
fig, axes = plt.subplots(num_rows, 2, figsize=(12, 6 * num_rows))

for column, ax in zip(numerical_columns, axes.flat):
    #Existing code for handling outliers and imputation
    sns.histplot(df[column], ax=ax, kde=True)

    max_value = (3 * df[column].std()) + df[column].mean()
    min_value = -((3 * df[column].std()) + df[column].mean())

    mean = df.loc[(df[column] > min_value) & (df[column] < max_value), column].mean()

    if not np.isnan(mean):
        df.loc[(df[column] < min_value) | (df[column] > max_value), column] = mean
        df[column].fillna(mean, inplace=True)
    else:
        df[column].fillna(df[column].mean(), inplace=True)

    ax.set_title(f"{column} Before & After", fontsize=15)

# Adjusting layout and display the subplots
plt.tight_layout()
plt.show()
```
